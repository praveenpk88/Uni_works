### Build a quantum circuit

```python
from qiskit import QuantumCircuit

# create a quantum circuit for 4 qubits
qc = QuantumCircuit(4)
qc.x(0) # apply a NOT to qubit 0
qc.h(1) # apply a Hadamard to qubit 1
qc.h(2) # apply a Hadamard to qubit 2
qc.cx(1, 3) # apply a CNOT to qubits 0 and 3
qc.measure_all() # measure all qubits
qc.draw(output='mpl') # draw the circuit
```

### Simulate the circuit classically

```python
# ---- Circuit Simulation: Sampling -----
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram

# initialise the sampler
sampler = StatevectorSampler()

# define the number of shots to sample
shots = 1024

# sample the results
result = sampler.run([qc], shots=shots).result()

# collect counts and plot histogram
counts = result[0].data.meas.get_counts()

# print the counts
print(f'Simulated string count: {counts}')
plot_histogram(counts)
```
