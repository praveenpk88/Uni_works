# Regular Expression Execises

In this exercise, we are going to work together and go through a few regex practice. You can refer to the following regex cheatsheet for convinence: [**Python Regex Cheatsheet**](https://www.debuggex.com/cheatsheet/regex/python). Note that we should not use `|` to simply concatenate the full text of all the matched cases. Otherwise, it will make no sense to use Regex. 

In this exercise, you can use the following function `test()` to check if your regex patterns fulfilled the requirements.

```python
import re

def test(pattern, matches, skips):
    count = 0
    
    for m in matches:
        if re.fullmatch(pattern, m):
            count +=1
            print( "\""+m+"\""," is matched successfully!")
        else:
            print ('Failed to match ',"\""+m+"\"")
    for s in skips:
        if not re.fullmatch(pattern, s):
            count += 1
            print( "\""+s+"\""," is skipped successfully!")
        else:
            print('Failed to skip ', "\""+s+"\"")
    print()
    if count == len(matches)+len(skips):
        print("Your defined pattern has fulfilled all requirements.")
    else:
        print("Your defined pattern has fulfilled",count, "out of", len(matches)+len(skips), "requirements.")
```

In this exercies, we not only try to match the text, we also want to capture data as indicated in the capture group. Therefore, we define another test function named `test_capture()` that can test whether we've captured the data properly.

```python
# define another test function, so that it tests whether we capture the capture group successfully too.
def test_capture(pattern, matches, skips,cap_groups = []):
    count = 0
    
    for i in range(0,len(matches)):
        result = re.fullmatch(pattern, matches[i])
        if result:
            count +=1
            print( "\""+matches[i]+"\""," is matched successfully!",end='\t')
            if cap_groups != [] and list(result.groups())!= cap_groups[i]: # the task requires capture groups
                count -=1
                print("But,",cap_groups[i]," are not captured properly")
            else: 
                print()
        else:
            print ('Failed to match ',"\""+matches[i]+"\"")
    for s in skips:
        if not re.fullmatch(pattern, s):
            count += 1
            print( "\""+s+"\""," is skipped successfully!")
        else:
            print('Failed to skip ', "\""+s+"\"")
    print()
    if count == len(matches)+len(skips):
        print("Your defined pattern has fulfilled all requirements.")
    else:
        print("Your defined pattern has fulfilled",count, "out of", len(matches)+len(skips), "requirements.")
    
```
